# 💻 Custom Tech Inventory Dataset

## 📌 Overview
This dataset contains information about 15 popular tech products across categories such as **Mobile, Laptop, Tablet, Audio, Wearable, Gaming, Camera, and Peripheral**.  
Each product is described with attributes including **Category, Price, Stock Availability, and Rating**.

---

## 🗂️ Dataset Structure
- **Product** → Name of the tech item  
- **Category** → Type (Mobile, Laptop, Tablet, etc.)  
- **Price** → Price in USD  
- **InStock** → Availability (True/False)  
- **Rating** → Average customer rating (out of 5)  
- **Index** → Custom ID format (`ID-001` to `ID-015`)

- **Highest Rated Products:** iPad Pro, Kindle Paperwhite, Logitech MX Master (4.9)

```python
import pandas as pd

# 1. Build the Dictionary (5 Columns, 15 Rows)
data = {
    'Product': ['iPhone 15', 'MacBook Air', 'Galaxy S23', 'iPad Pro', 'Dell XPS', 
                'Surface Pro', 'Pixel 8', 'AirPods Pro', 'Sony WH-1000XM5', 'Apple Watch',
                'Kindle Paperwhite', 'Nintendo Switch', 'GoPro Hero 12', 'Logitech MX Master', 'Razer Blade'],
    'Category': ['Mobile', 'Laptop', 'Mobile', 'Tablet', 'Laptop', 
                  'Tablet', 'Mobile', 'Audio', 'Audio', 'Wearable', 
                  'E-Reader', 'Gaming', 'Camera', 'Peripheral', 'Laptop'],
    'Price': [799, 999, 749, 1099, 1200, 899, 699, 249, 349, 399, 139, 299, 399, 99, 2399],
    'InStock': [True, True, False, True, False, True, True, True, True, False, True, True, True, True, False],
    'Rating': [4.7, 4.8, 4.5, 4.9, 4.4, 4.3, 4.6, 4.8, 4.7, 4.2, 4.9, 4.8, 4.5, 4.9, 4.6]
}

# 2. Create Custom Index
# We'll use a unique ID format like 'TECH-001', 'TECH-002', etc.
custom_indices = [f'ID-{i:03d}' for i in range(1, 16)]

# 3. Initialize the DataFrame
df_custom = pd.DataFrame(data, index=custom_indices)

# Display the result
print("--- Custom Tech Inventory Dataset ---")
print(df_custom)
```
